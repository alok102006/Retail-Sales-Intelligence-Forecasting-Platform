```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
```

```python
df.head()
```

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['month_name'] = df['order_date'].dt.month_name()
```

```python
df['sales'] = df['sales'].str.replace(',', '')
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
```

```python
import matplotlib.pyplot as plt

# Monthly Sales Trend
monthly_sales = df.groupby(df['order_date'].dt.to_period('M'))['sales'].sum()

plt.figure()
monthly_sales.plot()
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trend')
plt.show()


# Monthly Profit Trend
monthly_profit = df.groupby(df['order_date'].dt.to_period('M'))['profit'].sum()

plt.figure()
monthly_profit.plot()
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.title('Monthly Profit Trend')
plt.show()


# Annual Profit Trend
yearly_profit = df.groupby('year')['profit'].sum()

plt.figure()
yearly_profit.plot()
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.title('Annual Profit Trend')
plt.show()
```

```python
plt.figure()
df.groupby('discount')['profit'].mean().plot()
plt.xlabel('discount')
plt.ylabel('profit')
plt.show()
```

```python
df.plot.scatter(x='shipping_cost', y='profit')
```

```python
plt.figure()
df.groupby('category')['profit'].sum().plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.title('Category wise profit')
plt.show()
```

```python
plt.figure()
df.groupby('sub_category')['profit'].sum().plot(kind='bar')
plt.xlabel('sub_category')
plt.ylabel('Profit')
plt.title('sub_category wise profit')
plt.show()
```

```python
top_10_products = df.groupby('product_name')['profit'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6)) 
top_10_products.plot(kind='bar')

plt.xlabel('Product Name')
plt.ylabel('Total Profit')
plt.title('Top 10 Products by Profit')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout() 
plt.show()
```

```python
product_profits = df.groupby('product_name')['profit'].sum()


loss_making_products = product_profits[product_profits < 0].sort_values(ascending=True).head(10)


print("Top 10 Loss-Making Products:")
print(loss_making_products)


plt.figure(figsize=(10, 6))
loss_making_products.plot(kind='bar', color='red') 

plt.xlabel('Product Name')
plt.ylabel('Total Loss')
plt.title('Top 10 Loss-Making Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
monthly_sales = df.groupby(df['order_date'].dt.to_period('M'))['sales'].sum()
monthly_sales = monthly_sales.to_timestamp()
```

```python
monthly_sales = monthly_sales.reset_index()
monthly_sales['month_num'] = range(len(monthly_sales))
```

```python
from sklearn.linear_model import LinearRegression

X = monthly_sales[['month_num']]
y = monthly_sales['sales']

model = LinearRegression()
model.fit(X, y)
```

```python
future_months = 12
future_X = [[i] for i in range(len(monthly_sales), len(monthly_sales)+future_months)]

predictions = model.predict(future_X)
```

```python
monthly_sales['rolling_mean'] = monthly_sales['sales'].rolling(window=3).mean()



plt.plot(monthly_sales['order_date'], monthly_sales['sales'], label='Actual')
plt.plot(monthly_sales['order_date'], monthly_sales['rolling_mean'], label='Moving Avg')

plt.legend()
plt.show()
```

```python
monthly_sales = df.groupby(df['order_date'].dt.to_period('M'))['sales'].sum()
monthly_sales.index = monthly_sales.index.to_timestamp()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

model = ExponentialSmoothing(
    monthly_sales,
    trend='add',
    seasonal=None
).fit()

forecast = model.forecast(12)
```

```python
plt.plot(monthly_sales.index, monthly_sales.values, label='Actual')
plt.plot(forecast.index, forecast.values, label='Forecast')

plt.legend()
plt.show()
```
